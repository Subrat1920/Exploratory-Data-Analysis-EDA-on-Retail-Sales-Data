# TIME SERIES ANALYSIS

```python
retail_df.head()
```

```python
retail_df.set_index('Date', inplace=True)
```

```python
retail_df.head()
```

```python
retail_df.set_index('Date', inplace=True)
# to reset the index
```

```python
monthly_sales=retail_df['Total Amount'].resample('ME').sum()
monthly_sales
```

```python
plt.figure(figsize=(10,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Sales")
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()
```

```python
rolling_mean=monthly_sales.rolling(window=3).mean()
rolling_mean
```

```python
plt.figure(figsize=(10,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', label='Monthly Sales')
plt.plot(rolling_mean.index, rolling_mean.values, color='green', label='3-Months Rolling Mean')
plt.title("Monthly Sales with Rolling Mean")
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()
```

```python
menu_df.columns
```

```python
normal_value=['Calories', 'Calories from Fat','Total Fat', 'Saturated Fat', 'Trans Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']

melted_df = pd.melt(menu_df, id_vars=['Category'], value_vars=normal_value, var_name='Nutrient', value_name='Value')

plt.figure(figsize=(10,6))
sns.lineplot(data=melted_df, x='Nutrient', y='Value', hue='Category')
plt.title('Nutritional Values by Category')
plt.xlabel('Nutrient')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

```python
daily_percent_value=['Total Fat (% Daily Value)', 'Saturated Fat (% Daily Value)', 'Cholesterol (% Daily Value)', 'Sodium (% Daily Value)', 'Carbohydrates (% Daily Value)', 'Dietary Fiber (% Daily Value)', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)'] 

melted_df = pd.melt(menu_df, id_vars=['Category'], value_vars=daily_percent_value, var_name='Nutrient', value_name='Value')

plt.figure(figsize=(10,6))
sns.lineplot(data=melted_df, x='Nutrient', y='Value', hue='Category')
plt.title('Nutritional Percentage by Category')
plt.xlabel('Nutrient PErcentage Per Day')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```
